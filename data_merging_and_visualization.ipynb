## Missing Value Handling
- if a column has 30% or more missing in a column --> that column is unreliable-- remove the column
- else use imputation technique
- When preparing data for predictive modelling, do not impute in the target column using generic strategies.
- Imputation - using mean, median or mode, for time series --> forwardfill, backfill.
- You can choose granularity to come up with the value to impute

### Identify age groups based on titles

````python
titanic['titles'] = titanic.Name.str.split('[,.]').str[1].str.strip()
titanic['titles'].unique()
````

```python
young_males = ['Master']
young_females = ['Miss', 'Mlle']
adult_males = ['Mr', 'Don', 'Rev', 'Dr', 'Major', 'Sir',  'Col', 'Capt', 'Jonkheer']
adult_female = ['Mrs','Mme', 'Ms', 'Lady', 'the Countess']
```

````python
def age_category(title):
    if title in  ['Master']:  return 'young males'
    elif title in ['Miss', 'Mlle']: return 'young females'
    elif title in ['Mr', 'Don', 'Rev', 'Dr', 'Major', 'Sir',  'Col', 'Capt', 'Jonkheer']: return 'adult males'
    elif title in ['Mrs','Mme', 'Ms', 'Lady', 'the Countess']: return 'adult female'
    else : return "No Category"
````

````python
titanic['Age_category'] = titanic['titles'].apply(age_category)
titanic['Age_category']
````

````python
titanic.groupby('Age_category').agg({'Age': ['mean', lambda x: x.isna().sum()]})
````

````python
titanic.Age.isna().sum()
````

````python
titanic.Embarked.value_counts(dropna = False)
````

````python
titanic.Embarked.fillna('S').value_counts(dropna = False)
````

````python
age_grp = titanic.groupby('Age_category')['Age'].mean()
age_grp
````

````python
age_grp.loc['adult female']
````

````python
age_grp.index.tolist()
````

````python
for agecat in age_grp.index.tolist():
    titanic.loc[(titanic.Age_category == agecat) & (titanic.Age.isnull()), 'Age'] = age_grp[agecat]
````

````python
cities = pd.Series(['New York', 'New York', 'Sydney', 'New Delhi', 'New Delhi', 'Sydney', 'London', 'London'])
````

````python
cities
````

````python
from sklearn.preprocessing import LabelEncoder
````

````python
le = LabelEncoder()
le.fit(cities)
cities_le = pd.Series(le.transform(cities))
````

````python
pd.concat([cities, cities_le], axis = 1)
````

````python
travel = pd.Series(['Economy', 'Economy+','Economy','Economy','Economy', 'Business','Economy','Economy+', 'First Class', 'Business', 'Economy+'])
````

````python
travel
````

````python
le = LabelEncoder()
le.fit(travel)
travel_le = pd.Series(le.transform(travel))
pd.concat([travel, travel_le], axis = 1)
````

````python
from sklearn.preprocessing import OrdinalEncoder
````

````python
travel.values.reshape(-1,1)
````

````python
ode = OrdinalEncoder(categories = [['Economy', 'Economy+', 'Business','First Class' ]])
ode.fit(travel.values.reshape(-1,1))
travel_ode = pd.Series(ode.transform(travel.values.reshape(-1,1)).flatten())
````

````python
pd.concat([travel, travel_le, travel_ode], axis = 1)
````

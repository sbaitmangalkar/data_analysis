## 1. Creating Series

### 1.1 From List

````python
# Simple Series from a list
s = pd.Series([10, 20, 30, 40, 50])
print("Series from list:")
print(s)
print(f"\nType: {type(s)}")
print(f"Data type of values: {s.dtype}")
````

### 1.2 From List with Custom Index

````python
# simple list with custom index:
s = pd.Series([10, 20, 30, 40, 50], index = ['A', 'B', 'C','D', 'E'])
print("Series from list:")
print(s)
````

### 1.3 From Dictionary

````python
# Dictionary becomes Series with keys as index
prices = {'apple': 100, 'banana': 150, 'orange': 120, 'mango': 200}
s = pd.Series(prices)
print("Series from dictionary:")
print(s)
print(f"\nIndex comes from dictionary keys: {s.index.tolist()}")
````

### 1.4. Accessing Series Elements

````python
# usig label
print(f"Price for apple: Rs { s['apple']}")

# multiple elements
print(f"Price for apples & mangos: \n{ s[['apple', 'mango']]}")
````

````python
# usig label -> .loc
print(f"Price for apple: Rs { s.loc['apple']}")
````

````python
# using position .iloc
print("Element at position '2':")
print("Value: ", s.iloc[2])
print("Index: ",s.index[2])
````

### 1.5. Series Properties & Methods

````python
s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'], name='Prices')

print("Basic Properties:")
print(f"Index: {s.index.tolist()}")
print(f"Values: {s.values.tolist()}")
print(f"Data type: {s.dtype}")
print(f"Name: {s.name}")
print(f"Length: {s.size}")

print("\nStatistical Methods:")
print(f"Sum: {s.sum()}")
print(f"Mean: {s.mean():.2f}")
print(f"Median: {s.median():.2f}")
print(f"Min: {s.min()}")
print(f"Max: {s.max()}")
print(f"Std Dev: {s.std():.2f}")

print("\nDescribe (Summary Statistics):")
display(s.describe())

````

### 1.6. Series Operations

````python
prices = {'apple': 100, 'banana': 150, 'orange': 120, 'mango': 200}
prices_ser = pd.Series(prices)

discounts = {'apple': 10,  'orange': 10, 'mango': 5, 'grapes': 8}
discounts_ser = pd.Series(discounts)

prices_ser - discounts_ser
````

---
### Note: Series do not just support elementwise operation. Support index based operation.
---

````python
prices_ser.subtract(discounts_ser, fill_value = 0)
````

### 1.9. Custom Functions

````python
employees = pd.Series([5, 8, 9, 8.5, 7, 9.5], index = ['John', 'Kate', 'Prince', 'Amit', 'Nayan', 'Leah'], name = 'Performance Score')
employees
````

````python
res = []
for score in employees:
    if score >= 9 : r = 'promote & raise'
    elif score >= 7.5: r = 'raise'
    elif score >= 6 : r = 'review'
    else : r = 'monitor'
    res.append(r)
result = pd.Series(res, index = employees.index, dtype = str)
result
````

````python
# Wrapping it in a function
def res(score):
    if score >= 9 : return 'promote & raise'
    elif score >= 7.5: return 'raise'
    elif score >= 6 : return'review'
    else : return 'monitor'
result = employees.apply(res)
result
````

````python
result.str.title()
````

````python
result.index = result.index.str.upper()
result
````

### 2.13 Aggregation

````python
# Group by single column
print('Survival rate by class:')

df.groupby('Pclass').count()[['Survived']]
````

````python
df.groupby('Pclass')[['Survived']].count()
````

````python
df.groupby('Pclass')[['Survived']].agg(['count', 'sum', 'mean'])
````

````python
# for custom functions --> range
def percentage(x):
    return f'{round(x.mean() * 100,1)} %'
````

````python
df.groupby('Pclass')[['Survived']].agg(['count', 'sum', 'mean', percentage])
````

````python
df.groupby(['Pclass', 'Sex'])[['Survived']].agg(['count', 'sum', 'mean', percentage]).unstack()
````

````python
# pivot table
df.pivot_table(index = 'Pclass',
              columns = 'Sex',
              values = 'Survived',
              aggfunc = ['count', 'sum', 'mean', percentage] )
````

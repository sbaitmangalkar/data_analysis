```python
import matplotlib.pyplot as plt
```

```python
colors = ['aquamarine', 'gold', 'lightsteelblue', 'lightcoral', 'steelblue', 'plum', 'teal', 'cadetblue', 'powderblue', 'lightblue']
```

## Bar Chart

```python
vc = cars.Type.value_counts()

# create a figure
plt.figure(figsize=(10,5))

# plot the fig
plt.bar(x = vc.index, height = vc.values, color = colors)

# the axis labels
plt.xlabel('Types of Cars', size = 15, weight = 'bold', color = 'darkgreen')
plt.ylabel('Counts', size = 15, weight = 'bold', color = 'darkgreen')

# the axis limits
plt.ylim(0, 300)

# the axis ticks
plt.xticks(rotation = 0, size = 10, color = 'teal', weight = 'bold')
plt.yticks(range(0, 301, 75), rotation = 0, size = 12, color = 'teal', weight = 'bold')

# plot lines on existing plot
plt.plot([-0.92, 2], [150,150], color = 'red')
plt.plot([2, 2], [0,150], color = 'red')

# annotation
plt.annotate('coordinates: (2, 150)', xy = (2,150))

# annotate bars with their respective values
for i, val in enumerate(vc.index):
    plt.annotate(f"{vc[val]}".center(3), xy = (i - 0.15, vc[val]), size = 8, color = 'maroon', weight = 'bold')


# to manage spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.spines['left'].set_color('steelblue')
ax.spines['left'].set_linewidth(1.5)

ax.spines['bottom'].set_color('steelblue')
ax.spines['bottom'].set_linewidth(1.5)

# grid lines
plt.grid(axis = 'y', linestyle = '--', alpha = 0.3)

# the title
plt.title('Type of Cars Distribution', size = 20, weight = 'bold', color = 'darkblue', pad = 30)


# save the figure as png
plt.savefig('barchart.png')

# showing a fig
plt.show()
```

## Horizontal Bar Chart

```python
vc = cars.Type.value_counts(ascending=True)
plt.barh(y = vc.index, width = vc.values, color = colors)

# annotate the bars with their respective values

for i, val in enumerate(vc.index):
    plt.annotate(f"{vc[val]}".center(3), xy = (vc[val]+5, i-0.15), size = 10, color = 'maroon', weight = 'bold')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

## Pie Chart

```python
vc = cars.Type.value_counts(ascending = True)
vc.plot.pie()
plt.show()
```

```python
vc = cars.Type.value_counts()
plt.figure(figsize = (20,5))
plt.pie(x = vc.values, labels= vc.index,
    radius = 1.2,
    colors = colors,
    autopct='%.1f %%', 
    pctdistance= 0.5, labeldistance = 1,
    textprops= {'size' : 10, 'color': 'midnightblue', 'weight' : 'light' },
    wedgeprops= {'edgecolor': 'white'},
    # shadow = True,
    explode = [0.01, 0.02, 0.02, 0.03, 0.03, 0.5])

plt.title('Pie Chart', size = 20, pad = 25, color = 'darkgreen', weight = 'bold')

plt.savefig('Pie_chart_demo.png')

plt.show()
```

```python
# donut chart

vc = cars.Type.value_counts()
plt.figure(figsize = (20,5))
plt.pie(x = vc.values, labels= vc.index,
    radius = 1.2,
    colors = colors,
    autopct='%.1f %%', 
    pctdistance= 0.5, labeldistance = 1,
    textprops= {'size' : 10, 'color': 'midnightblue', 'weight' : 'light' },
    wedgeprops= {'edgecolor': 'white', 'width': 0.75},
    # shadow = True,
    )

plt.title('Donut Chart', size = 20, pad = 25, color = 'darkgreen', weight = 'bold')

plt.savefig('donut_chart.png')

plt.show()
```

## Bivariate: Categorical

### Make of Cars

```python
cars.Make.nunique()
```

#### Top 5 makes

```python
top_5_makes = cars.Make.value_counts().head(5).index
top_3_types = cars.Type.value_counts().head(3).index

subdata = cars[(cars.Make.isin(top_5_makes)) & (cars.Type.isin(top_3_types))]

ct = pd.crosstab(subdata.Make, subdata.Type)
ct
```

### Grouped Bar Chart

```python
ct.plot.bar(color=colors)
plt.show()
```

#### Stacked bar charts

```python
ct.plot.bar(color=colors, stacked = True)
plt.xticks(rotation = 0)
plt.legend(bbox_to_anchor = (1.3, 0.65))
plt.show()
```
